```python
import plotly.subplots as sp
import plotly.graph_objects as go
import plotly.express as px

fig = sp.make_subplots(rows=1, cols=11, shared_xaxes=True) 

fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['GR'], mode='lines', name='GR'), row=1, col=1)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['ROP5_RM'], mode='lines', name='ROP5_RM'), row=1, col=2)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['DT'], mode='lines', name='DT'), row=1, col=3)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['NPHI'], mode='lines', name='NPHI'), row=1, col=4)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['PHIF'], mode='lines', name='PHIF'), row=1, col=5)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['RT'], mode='lines', name='RT'), row=1, col=11)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['RHOB'], mode='lines', name='RHOB'), row=1, col=7)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['VSH'], mode='lines', name='VSH'), row=1, col=8)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['SW'], mode='lines', name='SW'), row=1, col=9)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['KLOGH'], mode='lines', name='KLOGH'), row=1, col=10)
fig.add_trace(go.Scatter(y=df['DEPTH'], x=df['pp_psi'], mode='lines', name='Pressure'), row=1, col=6)
for col_num, log_name in enumerate(['GR', 'ROP', 'DT', 'NPHI', 'PHIF', 'PP', 'RHOB', 'VSH', 'SW', 'KLOGH','RT']):
    fig.update_xaxes(title_text=log_name, row=1, col=col_num + 1)
# Update x-axis type for the 'RT' trace
fig.update_xaxes(type="log", row=1, col=11)
fig.update_xaxes(type="log", row=1, col=10)
fig.update_yaxes(autorange="reversed")
fig.update_layout(title_text='Subplots', template='plotly_dark', width=1920, height=900)

fig.show()

fig.write_html("subplot_plot.html")


```
